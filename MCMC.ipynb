```python
import numpy as np
import matplotlib.pyplot as plt
from skimage.io import imread, imsave
import time
np.random.seed(42)
```

```python
distribution = lambda x: (1+x**2/2)**(-2)
```

```python
n = 100000
s = np.zeros((n,1))
s[0] = 5
```

```python
time_start = time.time() #start time
for i in range(1,n):
    xprime = np.random.uniform(s[i-1]-0.5, s[i-1]+0.5)
    alpha = distribution(xprime) / distribution(s[i-1])
    if np.random.uniform() < alpha:
        s[i] = xprime
    else:
        s[i] = s[i-1]
```

```python
smin = min(s)
smax = max(s)

xx=np.linspace(smin, smax, n)
hist, bins = np.histogram(s, bins=xx.flatten(), density=True)
bins_center = 0.5*(bins[1:]+bins[:-1])
```

```python
plt.figure(figsize=(6,4))
plt.plot(bins_center, hist, label='hostogram of samples')
plt.plot(bins_center, distribution(bins_center), label='hostogram of samples')
plt.legend()
time_end = time.time()
time_c= time_end - time_start   #run time caculate
print('time cost', time_c, 's')
```
